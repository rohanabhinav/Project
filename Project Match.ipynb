```python
import pandas as pd
filePath = 'Match.csv'
def readPathCSV(filePath):
    return pd.read_csv(filePath)
df = readPathCSV(filePath)
print(df)
```

```python
df.groupby('Season_Year')['Match_SK'].count()
```

```python
import matplotlib.pyplot as plt
b = df[df['Season_Year']==2011].groupby('match_winner')['Country_id'].count()
q = b.index.tolist()
w = b.values.tolist()
plt.bar(q,w,color='red',width=0.6)
```
